```python
# Dependencies
import requests, json
import pandas as pd

# Google developer API key
from ba_config import g_key
```

```python
# (https://developers.google.com/places/web-service/search#TextSearchRequests)
my_phrase = "bakeries in jefferson square Topeka"
target_url = "https://maps.googleapis.com/maps/api/place/textsearch/json"

params = {
    "query": my_phrase,
    "key": g_key
    "radius"
}

response = requests.get(target_url, params)

# print the response URL, avoid doing for public GitHub repos in order to avoid exposing key
# print(response.url)
food_banks = response.json()
food_banks_json= json.dumps(food_banks, indent=4, sort_keys=True)
print(food_banks_json)
```

```python
names = []
addresses = []
lats = []
lngs = []
place_ids = []
ratings = []
price_levels = []
types = []

counter = 0
for food_bank in food_banks["results"]:
    names.append(food_bank["name"])
    addresses.append(food_bank["formatted_address"])
    lats.append(food_bank["geometry"]["location"]["lat"])
    lngs.append(food_bank["geometry"]["location"]["lng"])
    place_ids.append(food_bank["place_id"])
    ratings.append(food_bank["rating"])
    try:
        price_levels.append(food_bank["price_level"])
    except (KeyError, IndexError):
        price_levels.append(0)
    types.append(food_bank["types"])
    counter += 1

    
# create dataframe

    df = pd.DataFrame({
        "Name": names,
        "Address": addresses,
        "Latitude": lats,
        "Longitude": lngs,
        "Place ID": place_ids,
        "Rating": ratings,
        "Price Level": price_levels,
        "Type": types
    })
df
```
